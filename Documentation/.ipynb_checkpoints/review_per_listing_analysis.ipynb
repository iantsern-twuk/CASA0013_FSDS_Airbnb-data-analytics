# Change in Occupancy Analysis
This Notebook will use reviews per listing as a proxy to demonstrate the change in usage of Airbnb between 2023-2024 and 2022-2023.

**Stages to the workflow:**
1. Data processing
2. Calculations
3. Spatial autocorrelation and cluster (if any) identification

```python
#Loading packages
import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt 
import os
from scipy.stats import ttest_rel
import duckdb as db
from matplotlib import colors
import seaborn  # Graphics
import rioxarray  # Surface data manipulation
import xarray  # Surface data manipulation
from pysal.explore import esda  # Exploratory Spatial analytics
from pysal.lib import weights  # Spatial weights
import contextily  # Background tiles
from splot import esda as esdaplot
print(os.getcwd())
```

## 1) Data Processing

```python
#Load in both datasets
listings = pd.read_csv("data/clean/listings_provisonal.csv")
reviews = pd.read_csv("data/clean/reviews_provisional.csv")
```

```python
#Create a year column
reviews['date'] = pd.to_datetime(reviews["date"], format="%Y-%m-%d")
reviews['year'] = reviews.date.dt.year
#reviews.drop(columns = ['year_category'], inplace=True)
```

## 2) Reviews/Listing Calculation
### Cerys's method
1. Calculate total reviews for each listing per year
2. Get a count of listings for each ward
4. Get the sum of reviews for each ward
5. Divide sum of reviews by total number of listings
6. Calculate the difference between 22-23 and 23-24

```python
#Get location data
listings = gpd.GeoDataFrame(listings, geometry = gpd.points_from_xy(listings.longitude, 
                                                                    listings.latitude,
                                                                    crs ='epsg:4326'))

#Keep info for occupancy calculations
# listings = listings[["id", "geometry"]]
```

```python
#Step 1: Calculate total reviews for each listing
annual_reviews = reviews.groupby(['listing_id', 'year_category']).size().unstack(fill_value=0)
annual_reviews.rename(columns={'2022-2023': '2022_reviews', '2023-2024': '2023_reviews'}, inplace=True)
annual_reviews = annual_reviews.reset_index()
annual_reviews.columns.name = None

#Ensure annual_reviews is still a gdf
annual_reviews = annual_reviews.merge(listings, how='left', left_on='listing_id', right_on='id').drop(columns = ['id'])

annual_reviews['first_review'] = pd.to_datetime(annual_reviews['first_review'], format="%Y-%m-%d")

# calculate reviews per month for 2022
annual_reviews['2022_reviews_per_month'] = np.where(
    (pd.Timestamp('2022-09-10')-annual_reviews['first_review']).dt.days < 30,
    annual_reviews['2022_reviews'], # if days between first review and scrape < 30, use total # reviews
    np.where(
        annual_reviews['first_review'] > pd.Timestamp('2022-09-10'), # if the first review was after the last scrape date
        # reviews / # of months since first review
        annual_reviews['2022_reviews'] / (pd.Timestamp('2024-09-10')-annual_reviews['first_review']).dt.days / (365/12),  
        annual_reviews['2022_reviews'] / 12 # else num of reviews / 12
    )
)

# calculate reviews per month for 2023
annual_reviews['2023_reviews_per_month'] = np.where(
    (pd.Timestamp('2023-09-10')-annual_reviews['first_review']).dt.days < 30, 
    annual_reviews['2023_reviews'], # if days between first review and scrape < 30, use total # reviews
    np.where(
        annual_reviews['first_review'] > pd.Timestamp('2023-09-10'), # if the first review was after the last scrape date
        # reviews / # of months since first review
        annual_reviews['2023_reviews'] / (pd.Timestamp('2024-09-10')-annual_reviews['first_review']).dt.days / (365/12),  
        annual_reviews['2023_reviews'] / 12 # else # reviews / 12
    )
)

annual_reviews = gpd.GeoDataFrame(annual_reviews, geometry='geometry', crs=listings.crs)
```

```python
#Load in wards
wards_url = 'data/London_Ward_CityMerged.shp'
wards = gpd.read_file(wards_url)
columns_to_keep = ['NAME', 'GSS_CODE', 'geometry']
wards = wards.loc[:, columns_to_keep]
print(f'Wards projection: {wards.crs}')
print(f'\nListings projection: {annual_reviews.crs}')

#Transform reviews projection
annual_reviews = annual_reviews.to_crs(wards.crs)
print(f'\nNew listings projection: {annual_reviews.crs}')

wards.rename(columns={'GSS_CODE': 'ward_code', 'NAME': 'ward_name'}, inplace=True)
```

```python
# Step 2: Sum by wards
# Join listings to wards - within isn't working but intersects is! Why?
occupancy_by_ward = gpd.sjoin(wards, annual_reviews, how = 'left', predicate = 'intersects')

# Sum reviews per ward
ward_estimates = occupancy_by_ward.groupby('ward_code')[['2022_reviews_per_month', '2023_reviews_per_month']].sum()

ward_counts = occupancy_by_ward.groupby('ward_code')[['listing_id']].count()

# Add on ward name and geometry from previous dataframe
ward_estimates = ward_estimates.merge(wards, on='ward_code', how='left')
ward_estimates = ward_estimates.merge(ward_counts, on = 'ward_code', how = 'left')
ward_estimates = gpd.GeoDataFrame(ward_estimates, geometry='geometry', crs=annual_reviews.crs)
ward_estimates = ward_estimates.rename(columns = {'listing_id':'listing_count'})
```

```python
#Step 4: divide total ward reviews by total ward listing count
ward_estimates['2022_reviews_rate'] = ward_estimates['2022_reviews_per_month']/ward_estimates['listing_count']

ward_estimates['2023_reviews_rate'] = ward_estimates['2023_reviews_per_month']/ward_estimates['listing_count']

#Step 5: find the difference between reviews rate for 22-23 and 23-24
ward_estimates['reviews_rate_diff'] = ward_estimates['2023_reviews_rate']-ward_estimates['2022_reviews_rate']

# assign 0 to reviews_rate_diff if no listings in ward
ward_estimates['reviews_rate_diff'] = np.where(
    ward_estimates['listing_count'] == 0, 0, ward_estimates['reviews_rate_diff']
)
```

```python
ward_estimates.plot(column = "reviews_rate_diff", 
                    cmap="coolwarm", 
                    legend=True, 
                    scheme='user_defined', 
                    classification_kwds={'bins':[-2,-1,1,4]})
```

```python
# Generate W from the GeoDataFrame
w = weights.distance.KNN.from_dataframe(ward_estimates, k=5)
# Row-standardization
w.transform = "R"

ward_estimates["w_review_rate_diff"] = weights.lag_spatial(w, ward_estimates['reviews_rate_diff'])

ward_estimates["review_rate_diff_std"] = ward_estimates["reviews_rate_diff"] - ward_estimates["reviews_rate_diff"].mean()
ward_estimates["w_review_rate_diff_std"] = weights.lag_spatial(w, ward_estimates['review_rate_diff_std'])

lisa = esda.moran.Moran_Local(ward_estimates["reviews_rate_diff"], w)
```

```python
# Set up figure and axes
f, axs = plt.subplots(nrows=2, ncols=2, figsize=(12, 12))
# Make the axes accessible with single indexing
axs = axs.flatten()

# Subplot 1 #
# Choropleth of local statistics
# Grab first axis in the figure
ax = axs[0]
# Assign new column with local statistics on-the-fly
ward_estimates.assign(
    Is=lisa.Is
    # Plot choropleth of local statistics
).plot(
    column="Is",
    cmap="plasma",
    scheme="quantiles",
    k=5,
    edgecolor="white",
    linewidth=0.1,
    alpha=0.75,
    legend=True,
    ax=ax,
)

# Subplot 2 #
# Quadrant categories
# Grab second axis of local statistics
ax = axs[1]
# Plot Quadrant colors (note to ensure all polygons are assigned a
# quadrant, we "trick" the function by setting significance level to
# 1 so all observations are treated as "significant" and thus assigned
# a quadrant color
esdaplot.lisa_cluster(lisa, ward_estimates, p=1, ax=ax)

# Subplot 3 #
# Significance map
# Grab third axis of local statistics
ax = axs[2]
#
# Find out significant observations
labels = pd.Series(
    1 * (lisa.p_sim < 0.05),  # Assign 1 if significant, 0 otherwise
    index=ward_estimates.index  # Use the index in the original data
    # Recode 1 to "Significant and 0 to "Non-significant"
).map({1: "Significant", 0: "Non-Significant"})
# Assign labels to `db` on the fly
ward_estimates.assign(
    cl=labels
    # Plot choropleth of (non-)significant areas
).plot(
    column="cl",
    categorical=True,
    k=2,
    cmap="Paired",
    linewidth=0.1,
    edgecolor="white",
    legend=True,
    ax=ax,
)

# Subplot 4 #
# Cluster map
# Grab second axis of local statistics
ax = axs[3]
# Plot Quadrant colors In this case, we use a 5% significance
# level to select polygons as part of statistically significant
# clusters
esdaplot.lisa_cluster(lisa, ward_estimates, p=0.05, ax=ax)

# Figure styling #
# Set title to each subplot
for i, ax in enumerate(axs.flatten()):
    ax.set_axis_off()
    ax.set_title(
        [
            "Local Statistics",
            "Scatterplot Quadrant",
            "Statistical Significance",
            "Moran Cluster Map",
        ][i],
        y=0,
    )
# Tight layout to minimize in-between white space
f.tight_layout()

# Display the figure
plt.show()
```
