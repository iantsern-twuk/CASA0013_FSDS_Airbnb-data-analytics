```python
#count reviews by unique listing_id by year
df_reviews['date'] = pd.to_datetime(df_reviews['date'])
df_reviews['year'] = df_reviews['date'].dt.year

count_by_year = df_reviews.groupby(['listing_id', 'year']).size().reset_index(name='count')
sum_by_year = count_by_year.groupby('year')['count'].sum()
sum_2019to2024 = sum_by_year.loc[2019:2024].sum() #loc is to select the range of the year
sum_total = sum_by_year.sum()
count_raw = count_by_year.shape[0]

print(count_by_year.head())
print("______________________")
print(sum_by_year)
print("______________________")
print("sum_total =", sum_total)
print("sum_2019to2024 =", sum_2019to2024)
print("count_raw =", count_raw)
```

```python
#compare the review count and the number of Airbnb per year

df_reviews['date'] = pd.to_datetime(df_reviews['date'])
df_reviews['year'] = df_reviews['date'].dt.year
unique_ids_per_year = df_reviews.groupby('year')['listing_id'].nunique()

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(18, 6))

sum_by_year.plot(kind='bar', ax=axes[0])
axes[0].set_title('Count of Reviews in London from 2010 to 2024')
axes[0].set_xlabel('Year')
axes[0].set_ylabel('Total Count of Reviews')
axes[0].grid(True)
 
unique_ids_per_year.plot(kind='bar', ax=axes[1])
axes[1].set_title('Number of Airbnb per Year in London')
axes[1].set_xlabel('Year')
axes[1].set_ylabel('Number of Airbnb')
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

### 1. Beside year 2020-2021 the hotel industry suffered by covid-19, the number of Airbnb and the count of reviews are stable growing from 2010 to 2024.
### 2. the count of reviews in 2024 is less than 2023 because the data is only until 2024-09
### 3. After 2021, the number of Airbnb is growing faster than previous years (why?).


```python
avg_comment_per_airbnb_per_year = sum_by_year / unique_ids_per_year
selected_years = avg_comment_per_airbnb_per_year.loc[2019:2024]

plt.figure(figsize=(10, 6))
selected_years.plot(kind='bar') #how many kind do we have eg line, bar, scatter
plt.title('Average number of comments per Airbnb per year in London')
plt.xlabel('Year')
plt.ylabel('Average number of comments per Airbnb')
plt.grid(True)
plt.show()

selected_years
```

```python
#Try some boxplot to see the distribution of count of reviews
def filter_and_plot_seaborn(dataframe, quantile_value):

    threshold = dataframe['count'].quantile(quantile_value)
    filtered_data = dataframe[(dataframe['count'] <= threshold) & (dataframe['year'] >= 2019)]
    

    sns.set(style="whitegrid") #like darkgrid, whitegrid, dark, white, ticks
    
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='year', y='count', data=filtered_data)
    plt.title(f'Box Plot of Counts by Year (Filtered by {quantile_value*100:.0f}th Percentile)')
    plt.xlabel('Year')
    plt.ylabel('Count of reviews')
    plt.show()

filter_and_plot_seaborn(count_by_year, 0.95) #you can change the quantile value to see the different distribution
```

```python
#let's see the distribution of count of reviews from 2019 to 2024
year_a = 2019 
year_b = 2024    

df_reviews['date'] = pd.to_datetime(df_reviews['date'])
df_reviews['year'] = df_reviews['date'].dt.year
data_interval = df_reviews[(df_reviews['year'] >= year_a) & (df_reviews['year'] <= year_b)]
distribution_interval = data_interval['listing_id'].value_counts()

plt.figure(figsize=(10, 6))
distribution_interval.hist(bins=30)
plt.title(f'Distribution of reviews from {year_a} to {year_b}')
plt.xlabel('Count of reviews')
plt.ylabel('Number of Airbnb')
plt.grid(True)
plt.show()
```

### It's look like the long tail distribution, we might apply some method learn from QM.

```python
#Compare the distribution of count of reviews in each year
years = range(2019, 2025)

df_reviews['date'] = pd.to_datetime(df_reviews['date'])
df_reviews['year'] = df_reviews['date'].dt.year


fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(18, 12)) 
axes = axes.flatten() 

for i, year in enumerate(years):
    data_year = df_reviews[df_reviews['year'] == year]['listing_id'].value_counts()
    axes[i].hist(data_year, bins=30, alpha=0.7)
    axes[i].set_title(f'Distribution of reviews in {year}')
    axes[i].set_xlabel('Count of reviews')
    axes[i].set_ylabel('Number of Airbnb')
    axes[i].grid(True)

plt.tight_layout()
plt.show()
```

### The x and y axis are not standardized, we need to standardize them.

```python
#standardize the x and y axis

# range of year
years = range(2019, 2025)

# convert date to year
df_reviews['date'] = pd.to_datetime(df_reviews['date'])
df_reviews['year'] = df_reviews['date'].dt.year

# initialize the max value and the distribution list
max_reviews = 0
max_count = 0
distributions = []


for year in years:
    distribution = df_reviews[df_reviews['year'] == year]['listing_id'].value_counts()
    distributions.append(distribution)
    max_reviews = max(max_reviews, distribution.max())
    max_count = max(max_count, distribution.size)

fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(18, 12))
axes = axes.flatten()

for i, (year, distribution) in enumerate(zip(years, distributions)):
    # histogram
    axes[i].hist(distribution, bins=30, alpha=0.7, range=(0, max_reviews))
    axes[i].set_title(f'Distribution of reviews in {year}')
    axes[i].set_xlabel('Count of reviews')
    axes[i].set_ylabel('Number of Airbnb')
    axes[i].set_xlim(0, max_reviews)
    axes[i].set_ylim(0, max_count)
    axes[i].grid(True)

    # scatter plot
    counts = distribution.value_counts()
    reviews = counts.index
    frequencies = counts.values

    axes[i].scatter(reviews, frequencies,s=7, alpha=0.3, color='blue') 
    axes[i].set_xlabel('Count of reviews')
    axes[i].set_ylabel('Frequency of Airbnb listings')
    axes[i].set_xlim(0, max_reviews)
    axes[i].set_ylim(-200, max_count)
    axes[i].grid(True)

plt.tight_layout()
plt.show()
```

### There are more Airbnb in the market and also more and more Airbnb with low count of reviews in 2024 (but 2024 only has 9 months data).


```python
review_threshold = 2 #you can change the threshold to see the different distribution

df_reviews['date'] = pd.to_datetime(df_reviews['date'])
df_reviews['year'] = df_reviews['date'].dt.year

years = range(2011, 2025)

percentages = {}
for year in years:
    year_data = df_reviews[df_reviews['year'] == year]
    
    review_counts = year_data['listing_id'].value_counts()
    less_than_threshold = review_counts[review_counts <= review_threshold].count() #count the number of Airbnb with less than 2 reviews
    total_listings = review_counts.count()
    percentage = (less_than_threshold / total_listings) * 100
    percentages[year] = percentage

plt.figure(figsize=(10, 6))
plt.plot(list(percentages.keys()), list(percentages.values()), marker='o')
plt.title(f'Percentage of Airbnb listings with fewer or equal to {review_threshold} reviews per year')
plt.xlabel('Year')
plt.ylabel('Percentage')
plt.grid(True)
plt.show()
```

### In 2024, more than 30% of Airbnb in London had fewer or equal to 2 reviews, marking an increase of over 5% compared to 2023.

## Where is hot spots of reviews of Airbnb in London?


```python
count_by_year['listing_id'] = count_by_year['listing_id'].astype(int) #I join columns which I think are important for the analysis but we could join more columns.

columns_to_join = [
    'id', 'latitude', 'longitude', 'property_type', 'room_type',
    'has_availability', 'availability_30', 'availability_60', 'availability_90', 'availability_365',
    'host_id', 'host_name', 'amenities', 'price',
    'calculated_host_listings_count', 'calculated_host_listings_count_entire_homes',
    'calculated_host_listings_count_private_rooms', 'calculated_host_listings_count_shared_rooms',
    'reviews_per_month', 'review_scores_rating', 'review_scores_accuracy',
    'review_scores_cleanliness', 'review_scores_checkin', 'review_scores_communication',
    'review_scores_location', 'review_scores_value'
]

count_by_year_merged = count_by_year.merge(df_listings[columns_to_join], left_on='listing_id', right_on='id', how='left')
count_by_year_merged['year'] = pd.to_datetime(count_by_year_merged['year'].astype(str) + '-01-01 12:00:00') #convert the year to datetime
count_by_year_merged.head()

save_path = '/Users/tsernian/Documents/CASA/CASA0013_FSDS/CASA0013_FSDS_Airbnb_living la vida code-a/data/count_by_year_merged.csv'
count_by_year.to_csv(save_path, index=False, encoding='utf-8')

```

```python
#Apply H3 to count unique id in each hexagon of each year

def lat_lng_to_h3(lat, lng, resolution=10):
    return h3.latlng_to_cell(lat, lng, resolution)  

count_by_year_merged['h3_index'] = count_by_year_merged.apply(
    lambda row: lat_lng_to_h3(row['latitude'], row['longitude']), axis=1
)

unique_id_count = count_by_year_merged.groupby(['year', 'h3_index'])['id'].nunique().reset_index(name='unique_id_count')

print(unique_id_count.head())
print(unique_id_count.info(verbose=True))
save_path = '/Users/tsernian/Documents/CASA/CASA0013_FSDS/CASA0013_FSDS_Airbnb_living la vida code-a/data/unique_id_count.csv'
unique_id_count.to_csv(save_path, index=False, encoding='utf-8')

```
